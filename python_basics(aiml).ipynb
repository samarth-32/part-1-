```python
df['netsales']=df['Sales'] * df['Quantity']
df.head()
```

```python
df1=df.drop(['netsales'],axis=1)
df1.head()
```

```python
df1.columns
```

```python
#this is data subsetting; subset of data is put into another variable. 
#Double square brackets is used to achieve that

df2=df1[['Order ID','Order Date', 'Product ID','Category', 'Quantity', 'Discount','Sales','Profit']]
df2.head()
```

```python
#df3=df2.iloc[[0,2]-->no. of rows,: --> no of column, i.e. ':' indicates all columns

df3=df2.iloc[[0,2],:]
df3
```

```python
#sales more than 1000
high_sales = df2.loc[df2['Sales'] > 1000]
#variable=...loc[['column name]]
high_sales.describe()
#min starts from 1000
```

```python
#Scatter plot
import matplotlib.pyplot as p
p.scatter(df2['Sales'],df2['Profit'],c='red')
p.title('Sales vs Profit')
p.xlabel('Sales')
p.ylabel('Profit')
p.legend(['single element'])
```
